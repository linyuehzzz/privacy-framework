Read individual data

- Geographic level: tract
- Attributes: VOTINGAGE (2) $*$ HISPANIC (2) $*$ CENRACE (63)

```python
import pandas as pd
import numpy as np

filename_hist = 'data/franklin_hist.csv'
hist = pd.read_csv(filename_hist)

# block to tract
hist['TRACT'] = hist['GEOID10'].astype(str).str[:11]
col_names = hist.columns.to_numpy()
col_names = np.delete(col_names, [0, -1])
hist = hist.groupby('TRACT').sum()[col_names]
hist = hist.reset_index()
hist
```

```python
# HHGQ (8) $*$ VOTINGAGE (2) $*$ HISPANIC (2) $*$ CENRACE (63) to VOTINGAGE (2) $*$ HISPANIC (2) $*$ RACE (7)
n2, n3, n4 = 2, 2, 63

for y in range(n2):  # voting age
    y = '{number:0{width}d}'.format(width=2, number=y)
    col_names = [col for col in hist.columns if y in col[2:4] and len(col)==8]

    for z in range(n3):  # ethnicity
        z = '{number:0{width}d}'.format(width=2, number=z)
        col_names2 = [col for col in col_names if z in col[4:6]]

        col_two_or_more_races = []
        for x in range(n4):  # race
            if x >= 0 and x <= 5:
                x = '{number:0{width}d}'.format(width=2, number=x)
                col_names3 = [col for col in col_names2 if x in col[6:8]]
                hist[x + y + z] = hist[col_names3].sum(axis=1)
            else:
                x = '{number:0{width}d}'.format(width=2, number=x)
                col_names3 = [col for col in col_names2 if x in col[6:8]]
                col_two_or_more_races.extend(col_names3)
        hist['06' + y + z] = hist[col_two_or_more_races].sum(axis=1)

hist.drop([col for col in hist.columns if len(col)==8], axis=1, inplace=True)
hist            
```

```python
## prioritize privacy
import pandas as pd
import numpy as np
from gurobipy import Model, GRB, quicksum, abs_

# initialize model
m1 = Model('td')
m1.params.NonConvex = 2
env0 = m1.getMultiobjEnv(0)
env1 = m1.getMultiobjEnv(1)
env0.setParam('TimeLimit', 5000)
env1.setParam('TimeLimit', 5000)
# m1.Params.LogToConsole = 0

# add decision variables and objective function
theta, A_tilde, b, d, md, obj1, obj2 = {}, {}, {}, {}, {}, {}, {}
for k in range(K):
    for i in V[k]:
        for j in range(I):
            # decision variables
            theta[i, j, k] = m1.addVar(vtype=GRB.CONTINUOUS, lb=0, ub=1, name="theta_%d_%d_%d"%(i, j, k))

eps = np.finfo(np.float64).eps
for k in range(K):
    for j in range(I):
        A_tilde[j, k] = m1.addVar(vtype=GRB.CONTINUOUS, lb=0, ub=float('inf'), name="A_tilde_%d_%d"%(j, k))
        obj1[j, k] = m1.addVar(vtype=GRB.CONTINUOUS, lb=0, ub=1, name="obj1_%d_%d"%(j, k))
        if j in V[k]:
            m1.addConstr(A_tilde[j, k] == A[j, k] * theta[j, j, k])
            m1.addConstr(obj1[j, k] * A_tilde[j, k] == theta[j, j, k])
        else:
            m1.addConstr(A_tilde[j, k] == A[j, k] + quicksum(T[i, j, k] * theta[i, j, k] * A[i, k] for i in V[k]))
            b[j, k] = m1.addVar(vtype=GRB.BINARY, name="b_%d_%d"%(j, k))
            m1.addConstr((b[j, k] == 0) >> (A_tilde[j, k] == 0))
            m1.addConstr((b[j, k] == 1) >> (A_tilde[j, k] >= eps))
            m1.addConstr(obj1[j, k] * A_tilde[j, k] == b[j, k])
        
        d[j, k] = m1.addVar(vtype=GRB.CONTINUOUS, lb=-float('inf'), ub=float('inf'), name="d_%d_%d"%(j, k))
        m1.addConstr(d[j, k] == A[j, k] - A_tilde[j, k])
        md[j, k] = m1.addVar(vtype=GRB.CONTINUOUS, lb=0, ub=float('inf'), name="md_%d_%d"%(j, k))
        m1.addConstr(md[j, k] == abs_(d[j, k]))
        obj2[j, k] = m1.addVar(vtype=GRB.CONTINUOUS, lb=0, ub=1, name="obj2_%d_%d"%(j, k))
        m1.addConstr(obj2[j, k] * (A[j, k] + A_tilde[j, k]) == md[j, k])

# m1.setObjective(quicksum(quicksum(obj1[j, k] for j in range(I)) for k in range(K)), GRB.MINIMIZE)
m1.setObjectiveN(quicksum(quicksum(obj1[j, k] for j in range(I)) for k in range(K)) / (K * I), index=0, priority=2, name="privacy")
m1.setObjectiveN(quicksum(quicksum(obj2[j, k] for j in range(I)) for k in range(K)) / (K * I), index=1, priority=1, name="utility")

for k in range(K):
    if len(V[k]) == 0:
        continue
    for i in V[k]:
        # add constraints
        m1.addConstr(quicksum(theta[i, j, k] for j in range(I)) == 1)
        m1.addConstr(theta[i, i, k] + quicksum(T[i, j, k] * theta[i, j, k] for j in range(I)) == 1)      
        
for j in range(I):
    m1.addConstr(quicksum(quicksum(T[i, j, k] * theta[i, j, k] * A[i, k] for i in V[k]) for k in range(K)) <= nj)

m1.update()
m1.optimize()
```
